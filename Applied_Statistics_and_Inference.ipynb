**Question 1 : What are Type I and Type II errors in hypothesis testing, and how do they
impact decision-making?**

  - Type I Error and Type II Error: 10 Differences, ExamplesType I and Type II errors are incorrect conclusions in hypothesis testing: a Type I error is rejecting a true null hypothesis (a false positive), while a Type II error is failing to reject a false null hypothesis (a false negative). These errors impact decision-making by forcing a trade-off between them; minimizing one often increases the other. The severity of each error depends on the context, such as whether it's more costly to convict an innocent person (Type I) or let a guilty one go free (Type II).

#Impact on decision-making

**Trade-off:** There is an inherent trade-off between Type I and Type II errors. You can decrease the probability of one type of error, but only by increasing the probability of the other.

**Significance Level (\(\alpha \)):** The significance level (\(\alpha \)) is the probability of making a Type I error. A lower \(\alpha \) reduces the chance of a false positive but increases the chance of a false negative (Type II error).

**Statistical Power (\(1-\beta \)):** Statistical power is the probability of correctly rejecting a false null hypothesis. It is the complement of the Type II error rate (\(\beta \)). Increasing sample size or statistical power can help reduce both types of errors, but the trade-off still exists.

**Context is key**: The decision on which error to minimize depends on the real-world consequences of each mistake.For example, in a medical trial

  - for an expensive new drug, a Type I error would be incorrectly concluding the new drug is more effective than the cheaper one, which would cost patients more for no benefit.
  
   - In a criminal trial, a Type I error is convicting an innocent person, while a Type II error is failing to convict a guilty person. The legal system prioritizes avoiding the Type I error.
   
   **Balancing risk:** Decision-makers must balance the costs of both types of errors. Sometimes, it is more critical to avoid a false positive (like the drug example), while other times, it is more critical to avoid a false negative (like missing a real medical effect).

**Question 2:What is the P-value in hypothesis testing, and how should it be interpreted
in the context of the null hypothesis?**

  - A p-value is the probability of observing results as extreme as, or more extreme than, the results of your sample, assuming the null hypothesis is true. In the context of the null hypothesis (\(H_{0}\)), a small p-value (typically \(\le 0.05\)) suggests there is strong evidence against the null hypothesis, leading you to reject it. A large p-value (e.g., \(>0.05\)) indicates weak evidence against the null hypothesis, so you fail to reject it.
  
#Interpretation in the context of the null hypothesis
  
 **Small p-value (e.g., \(p\le 0.05\)):**

 - Indicates the observed data is unlikely to have occurred by random chance alone if the null hypothesis were true
 - Action: Reject the null hypothesis in favor of the alternative hypothesis.

**Example:** A p-value of \(0.03\) for a new drug test means there is a \(3\%\) chance of getting the observed results if the drug had no real effect (i.e., the null hypothesis is true).

**Large p-value (e.g., \(p>0.05\)):**

  - Indicates the observed data is consistent with what would be expected if the null hypothesis were true.

  - Action: Do not reject the null hypothesis. There is not enough evidence to conclude the alternative is true.
  
 **Example:** A p-value of \(0.44\) means there is a \(44.2\%\) chance of getting the results observed purely by chance, which is not strong enough evidence to reject the null hypothesis that the two groups are the same.
  
  **Important note:** The p-value is not the probability that the null hypothesis is true or false; it is a measure of how incompatible your data is with the null hypothesis.

**Question 3:Explain the difference between a Z-test and a T-test, including when to use each?**

  - A Z-test is used when the population standard deviation is known and the sample size is large (\(n\ge 30\)), while a t-test is used when the population standard deviation is unknown and the sample size is small (\(n<30\)).Both are hypothesis tests, but the t-test uses a t-distribution that accounts for the extra uncertainty from an unknown population standard deviation, while the Z-test uses a standard normal distribution.
  
#When to use each
  **Use a Z-test when**
  
  You know the standard deviation of the entire population you are studying, and your sample is large (\(n\ge 30\)).You are comparing the means of two populations and both population standard deviations are known, or you have a large enough sample size for both.
  
  **Use a T-test when:**
  
  The population standard deviation is unknown and must be estimated from the sample data.Your sample size is small (\(n<30\)). The t-distribution is more appropriate here because the smaller sample size introduces more uncertainty. For larger sample sizes (\(n\ge 30\)), a t-test will yield similar results to a Z-test.You are conducting a paired-samples t-test to compare the means of two related groups, such as measuring the same individuals before and after a treatment.

**Question 4:What is a confidence interval, and how does the margin of error influence its width and interpretation?**
?
  - A confidence interval provides a range of plausible values for an unknown population parameter, such as a population mean or proportion, based on data collected from a sample [3, 4]. It is typically expressed as "We are X% confident that the true parameter falls within this interval," where X% is the confidence level (commonly 90%, 95%, or 99%) [4].
  
  The margin of error (MOE) directly influences both the width and the interpretation of the confidence interval:
  
  **Influence on Width:** The confidence interval is calculated by taking the sample statistic (e.g., sample mean) and adding and subtracting the margin of error (Confidence Interval = Sample Statistic \(\pm \) Margin of Error) [1, 2].
  - A larger margin of error results in a wider confidence interval [1, 3].
  - A smaller margin of error results in a narrower confidence interval [1, 3].
  - A wider interval suggests less precision in estimating the true population parameter [3].
  
  **Influence on Interpretation:** The margin of error quantifies the maximum expected difference between the sample statistic and the actual population parameter [2, 3]
  - A larger margin of error indicates a higher degree of uncertainty in the estimate [3].
  - A smaller margin of error indicates a higher degree of precision and confidence in the sample statistic as a representation of the population [3].
  
  In summary, a larger margin of error leads to a wider, less precise confidence interval, while a smaller margin of error leads to a narrower, more precise interval [1, 3]. The choice of confidence level also affects the margin of error; for a given sample size, a higher confidence level (e.g., 99% vs. 95%) requires a larger margin of error to be more certain that the interval captures the true parameter [3, 4].


**Question 5: Describe the purpose and assumptions of an ANOVA test. How does it extend hypothesis testing to more than two groups?**

  - ANOVA's purpose is to determine if there are significant differences among the means of three or more groups by analyzing variances within and between groups. It extends hypothesis testing to multiple groups by using an F-test, which compares the variance "between" the groups to the variance "within" the groups; a large F-ratio suggests that the difference between group means is too large to be due to chance. Key assumptions for using ANOVA include that samples are random and independent, data within each group is normally distributed, and the groups have equal variances (homoscedasticity).
#Purpose of ANOVA
  
 **Compares multiple means:** ANOVA determines if there is a statistically significant difference between the means of three or more independent groups.

 **Analyzes variance:** It works by partitioning the total variance in the data into two components: the variance between the groups and the variance within the groups.

**Tests a hypothesis:** The null hypothesis (\(H_{0}\)) is that all group means are equal, while the alternative hypothesis (\(H_{a}\)) is that at least one group mean is different.

#How it extends hypothesis testing to more than two groups

**Avoids multiple t-tests:** Instead of performing multiple t-tests (which would inflate the Type I error rate), ANOVA performs a single test to compare all group means simultaneously.

**Uses the F-statistic:** ANOVA calculates an F-statistic, which is the ratio of the variance between groups to the variance within groups (\(F=\frac{Variance_{between}}{Variance_{within}}\)).

**Determines significance:** A large F-statistic, with a large variation between groups compared to the variation within groups, indicates that the differences between the group means are statistically significant, leading to the rejection of the null hypothesis.

#Assumptions of ANOVA

**Independence:** Observations within each group and across groups are independent of one another.

**Normality:** The data for each group is drawn from a normally distributed population.

**Homogeneity of variances (Homoscedasticity):** The variances of the populations from which the samples are drawn are equal.

    """

        pop_mean (float): The assumed population mean (H0).
        pop_std (float): The known population standard deviation.

    sample_data = np.array(data)

#**Question 9: Write a Python function to calculate and visualize the confidence interval for a sample mean**


```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_and_visualize_confidence_interval(data, confidence_level=0.95, known_pop_std=None):
    """
    Calculates and visualizes the confidence interval for a sample mean.

    Args:
        data (list or np.array): The sample data.
        confidence_level (float): The desired confidence level (e.g., 0.95 for 95%).
        known_pop_std (float, optional): Known population standard deviation. If provided,
                                         a Z-interval is calculated. Otherwise, a T-interval is used.

    Returns:
        tuple: A tuple containing the sample mean, margin of error, and the confidence interval.
    """
    sample_data = np.array(data)
    n = len(sample_data)
    sample_mean = np.mean(sample_data)

    if n < 2:
        print("Error: Sample size must be at least 2 to calculate confidence interval.")
        return None, None, None

    if known_pop_std:
        # Z-interval (assumes population standard deviation is known)
        std_error = known_pop_std / np.sqrt(n)
        z_critical = stats.norm.ppf(1 - (1 - confidence_level) / 2)
        margin_of_error = z_critical * std_error
        method = "Z-interval"
    else:
        # T-interval (assumes population standard deviation is unknown)
        sample_std = np.std(sample_data, ddof=1) # Use ddof=1 for sample standard deviation
        std_error = sample_std / np.sqrt(n)
        degrees_freedom = n - 1
        t_critical = stats.t.ppf(1 - (1 - confidence_level) / 2, degrees_freedom)
        margin_of_error = t_critical * std_error
        method = "T-interval"

    confidence_interval = (sample_mean - margin_of_error, sample_mean + margin_of_error)

    print(f"Method: {method}")
    print(f"Sample Size (n): {n}")
    print(f"Sample Mean: {sample_mean:.4f}")
    print(f"Standard Error: {std_error:.4f}")
    print(f"Margin of Error: {margin_of_error:.4f}")
    print(f"Confidence Level: {confidence_level*100:.0f}%")
    print(f"Confidence Interval: ({confidence_interval[0]:.4f}, {confidence_interval[1]:.4f})")

    # Visualization
    plt.figure(figsize=(10, 4))
    plt.errorbar(x=sample_mean, y=0, xerr=margin_of_error, fmt='o', color='blue', capsize=5, label='Confidence Interval')
    plt.axvline(x=sample_mean, color='red', linestyle='--', label='Sample Mean')
    plt.title(f'{int(confidence_level*100)}% Confidence Interval for Sample Mean ({method})')
    plt.xlabel('Value')
    plt.yticks([]) # Hide y-axis ticks
    plt.legend()
    plt.grid(axis='x', linestyle=':', alpha=0.7)
    plt.xlim(sample_mean - margin_of_error * 2, sample_mean + margin_of_error * 2) # Adjust x-axis limits
    plt.show()

    return sample_mean, margin_of_error, confidence_interval

# --- Example Usage ---
# 1. Using a T-interval (population standard deviation unknown)
print("\n--- Example 1: T-interval (Unknown Pop Std Dev) ---")
sample_data_1 = [25, 28, 22, 27, 26, 23, 29, 24, 26, 25, 30, 21, 28, 26, 27]
mean_1, moe_1, ci_1 = calculate_and_visualize_confidence_interval(sample_data_1, confidence_level=0.95)

# 2. Using a Z-interval (population standard deviation known)
# Let's use the binomial_data from Question 7 as our "population" and draw a sample from it.
# Assume we know the standard deviation of this larger 'population'.
if 'binomial_data' in globals():
    print("\n--- Example 2: Z-interval (Known Pop Std Dev - using binomial_data) ---")
    # Take a sample from the binomial_data
    np.random.seed(42) # for reproducibility
    sample_data_2 = np.random.choice(binomial_data, size=50, replace=True) # Larger sample for CLT effect

    # For this example, let's assume we know the *true* standard deviation of the underlying binomial process (or a large enough population).
    # For a binomial distribution, std dev = sqrt(n_trials * p * (1-p)).
    # However, here 'binomial_data' represents observations of successes, not the underlying trials.
    # So we'll simulate a known population std based on the binomial_data itself for demonstration purposes.
    known_std_for_example = np.std(binomial_data) # This is an idealization for demonstration

    mean_2, moe_2, ci_2 = calculate_and_visualize_confidence_interval(sample_data_2, confidence_level=0.99, known_pop_std=known_std_for_example)
else:
    print("\n--- Example 2 Skipped: binomial_data not found. Please run Question 7's cell first. ---")
```

#**Question 10: Perform a Chi-square goodness-of-fit test using Python to compare observed and expected distributions, and explain the outcome.**


```python
import numpy as np
from scipy.stats import chisquare

# --- Define Observed and Expected Frequencies ---

# Scenario: A coin is tossed 100 times. We want to test if it's a fair coin.
# Expected distribution for a fair coin: 50 heads, 50 tails.

observed_frequencies = np.array([58, 42]) # Example: 58 heads, 42 tails observed
expected_frequencies = np.array([50, 50]) # Expected: 50 heads, 50 tails for a fair coin

# Another scenario: Rolling a six-sided die 60 times. Is the die fair?
# Expected: 10 occurrences for each face (60 / 6 = 10).
# observed_frequencies = np.array([8, 12, 10, 11, 9, 10])
# expected_frequencies = np.array([10, 10, 10, 10, 10, 10])

# --- Perform the Chi-square Goodness-of-Fit Test ---

# chisquare returns the chi-square statistic and the p-value
chi_square_statistic, p_value = chisquare(f_obs=observed_frequencies, f_exp=expected_frequencies)

# --- Interpret the Outcome ---
alpha = 0.05 # Significance level

print(f"Observed Frequencies: {observed_frequencies}")
print(f"Expected Frequencies: {expected_frequencies}")
print(f"Chi-square Statistic: {chi_square_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
print(f"Significance Level (alpha): {alpha}")

if p_value < alpha:
    interpretation = "Reject the null hypothesis (H0). There is a statistically significant difference between the observed and expected distributions."
else:
    interpretation = "Fail to reject the null hypothesis (H0). There is no statistically significant difference between the observed and expected distributions."

print(f"\nOutcome: {interpretation}")

print("\nExplanation of Outcome:")
print("The null hypothesis (H0) for a Chi-square goodness-of-fit test is that the observed data's distribution matches the expected distribution. The alternative hypothesis (H1) is that they are different.")
print("A small p-value (typically < alpha, e.g., 0.05) suggests that the observed deviation from the expected distribution is unlikely to have occurred by chance, leading us to reject H0.")
print("A large p-value (typically >= alpha) suggests that the observed differences could easily be due to random chance, and thus we fail to reject H0, meaning we don't have enough evidence to claim the distributions are different.")

if p_value < alpha:
    print(f"In this case, since the p-value ({p_value:.4f}) is less than alpha ({alpha}), we conclude that the observed coin toss results are significantly different from what we would expect from a fair coin.")
else:
    print(f"In this case, since the p-value ({p_value:.4f}) is not less than alpha ({alpha}), we conclude that there is not enough evidence to say that the observed coin toss results are significantly different from what we would expect from a fair coin. The coin appears to be fair (or we don't have enough evidence to say otherwise).")
```
